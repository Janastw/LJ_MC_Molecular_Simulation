```python
import math
import matplotlib.pyplot as plt

# create a function to calculate Lennard Jones interaction energy for an incremental range of distances, epsilon term and sigma term
def calculate_LJ(r_min, r_max,r_inc,o,e):
    """
    The Lennard Jones interation between two particles

    Parameters
    ----------
    r_min : integer
        The minimum r to calculate. The function will divide this value by 10.

    r_max : integer
        The maximum r to calculate. The function will divid this value by 10.

    r_inc : integer
        The incremental increase in r to run calculations. The function will divide this value by 10.

    r : float
        The distance between the two particles in reduced units

    o : float
        The distance at which the particle-particle potential enerty is zero.

    e : float
        The depth of the potential well

    Returns
    -------
    x_values : List
        List of distances used to run calculations
        
    y_values : List
        List of the resulting values of pairwise energy between two particles in reduced units
    """
    # create empty lists to store distances and pairwise energy values for plotting
    x_values = []
    y_values = []
    
    # create for loop to calculate pairwise energies over a range of distances
    for i in range(r_min,r_max,r_inc):
        r = i/10
        r6_term = math.pow((o/r),6)
        r12_term = math.pow(r6_term,2)
        pairwise_energy = 4*e*(r12_term - r6_term)
        
        # append values to lists for plotting
        x_values.append(r)
        y_values.append(pairwise_energy)
    return x_values, y_values
```

```python
# create three datasets of distances and energies for plotting
x1, y1 = calculate_LJ(1,51,1,1,1)
x2, y2 = calculate_LJ(1,51,1,2,1)
x3, y3 = calculate_LJ(1,51,1,1,2)

# plot all 3 datasets on the same graph
plt.scatter(x1,y1, color='blue', label = 'sigma = 1, epsilon = 1')
plt.scatter(x2,y2, color='red', label = 'sigma = 2, epsilon = 1')
plt.scatter(x3,y3, color='orange', label = 'sigma = 1, epsilon = 2')
plt.ylim(-3,8)
plt.xlabel("Distance Between Two Particles")
plt.ylabel("Pairwise Energy")
plt.legend()
plt.show()
```

## Reflection

### Approach
The first step taken was to create a function where a user can establish an incremental range of distances and calculate pairwise energy values for a given epsilon and sigma. The `range` function was used in the `for` loop so the inputs need to be integers that get scaled back to the necessary values. The resulting values of distances and energy values in the function were stored as lists that were returned. From there, the function is called three consecutive times, and the resulting 3 datasets are plotted on the graph.

### Observations
When r is less than sigma, the energy is steeply positive as r approaches 0. When r is greater than sigma, the energy is negative but approaches 0 as r increases. The value of epsilon lowers the energy at the optimal distance.
