# Calculation of $\pi$

This notebook will use Monte Carlo integration to calculate the value of $\pi$ using a unit circle.

PRocedure:

1. Start with a count inside of the circle = 0.
2. Generate a random point on the range (0,1).
3. Determine if the point lies within the unit circle.
4. If inside the circle, increase the counter.
5. Repeat steps 2-4 for desired number of samples.
6. Calculate the ratio of points under the curve vs. the total number of points.

```python
import math
import random
```

```python
# define total number points I will generate
n_samples = 100

# define variable for count
num_inside = 0

# Create a for loop for generating n_samples number of points
for i in range(n_samples):

    # Generate a random point between 0 adn 1 for our x value
    x = random.random()
    
    # Generate a random point between 0 and 1 for our y value
    y = random.random()

    # Calculate the distance of the point from the origin
    distance = math.sqrt( x**2 + y**2)

    # Increase the count if distance is less than 1.
    if distance < 1:
        num_inside += 1

# Calculate integral
integral = num_inside / n_samples

my_pi = 4 * integral

print(my_pi)
```

```python
# visualization using matplotlib
# matplotlib is the only library not in the PSL that you can use.
import matplotlib.pyplot as plt

# create a figure
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(111)


# define total number points I will generate
n_samples = 1000

# define variable for count
num_inside = 0

# Create a for loop for generating n_samples number of points
for i in range(n_samples):

    # Generate a random point between 0 adn 1 for our x value
    x = random.random()
    
    # Generate a random point between 0 and 1 for our y value
    y = random.random()

    # Calculate the distance of the point from the origin
    distance = math.sqrt( x**2 + y**2)

    # Increase the count if distance is less than 1.
    if distance < 1:
        num_inside += 1
        ax.plot(x, y, "ob")
    else:
        ax.plot(x, y, "r*")    


# Calculate integral
integral = num_inside / n_samples

my_pi = 4 * integral

print(my_pi)
```
