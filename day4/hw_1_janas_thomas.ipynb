Estimating $pi$ by evaluating integral using Monte Carlo Integration

```python
import math
import random
import matplotlib.pyplot as plt

# Generate points!
total_points = 1000
points_under_curve = 0

fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(111)
counter = 1

for i in range(total_points):
    x = random.random() * (counter)
    y = random.random()
    if counter == 1:
        counter = -1
    elif counter == -1:
        counter = 1

    if y < (1 / (1 + x**2)):
        points_under_curve += 1
        ax.plot(x, y, "ob")
    else:
        ax.plot(x, y, "r*")
    


monte_carlo_calculation = points_under_curve / total_points

print(monte_carlo_calculation)
print(math.pi / 4)



```
