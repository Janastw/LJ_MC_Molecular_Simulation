```python
import math

# Generates the increments of energies from -2 to 2
energies = []
for x in range(-200, 201, 1):
    energies.append(x / 100)

def calculate_probability(energies, temp):
    """
    Calculates the probability based on delta_e's and a specific temperature

    Parameters
    ----------
    energies: list
        Range of values of energy (delta_e).
    temp: float
        Specific temperature to calculate beta
    
    Returns
    -------
    probabilities: list
        list of probabilities of acceptance corresponding to each energy in the parameters
    """

    probabilities = []

    # if x in energies < 2 and > 2:
    for delta_e in energies:
        if delta_e <= 0.0:
            probability = 100
        else:
            p_acc = math.exp(-delta_e * (1 / temp))
            # Probability = p_acc
            probability = p_acc * 100
        probabilities.append(probability)
    
    return probabilities
```

```python
import matplotlib.pyplot as plt

temp1_probabilities = calculate_probability(energies, 0.4)
temp2_probabilities = calculate_probability(energies, 0.9)
temp3_probabilities = calculate_probability(energies, 1.4)



plt.plot(energies, temp1_probabilities, label='0.4')
plt.plot(energies, temp2_probabilities, label='0.9')
plt.plot(energies, temp3_probabilities, label='1.4')


plt.xlabel('Energies')
plt.ylabel('Probability of Acceptance (%)')

plt.legend()
```
